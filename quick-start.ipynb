# Sample Data
## Dataset 1: Titanic data

```python
import pandas as pd
print(pd.__version__)
titanic = pd.read_csv('./data/titanic.csv')
titanic.head()
```

## Dataset 2 :vaccine

support `date` columns

```python
import pandas as pd
vaccine = pd.read_csv('./data/country_vaccinations.csv')
vaccine['date'] = pd.to_datetime(vaccine['date'])
vaccine.head()
```

# summarytools
## Out-of-Box Summary

```python
from summarytools import dfSummary
dfSummary(titanic)
```

## Collapsible Summary

```python
from IPython.display import display
display(dfSummary(titanic, is_collapsible=True))
display(dfSummary(vaccine, is_collapsible=True))
```

## Tabbed Summary

```python
from summarytools import tabset
tabset({
    'titanic': dfSummary(titanic).to_html(),
    'vaccine': dfSummary(vaccine).to_html()
    })
```

# For very wide tables

use `num_pro` to parallel processing the columns. ( available since v0.3.0 )

```python
# concatenate `titanic` dataset 100x to make it super wide.
df_wide = titanic.copy()
for i in range(100):
    df_wide = pd.concat([df_wide, titanic.rename(lambda x: f'{x}_{i:04d}', axis='columns')], axis = 1)
print(df_wide.shape)
```

```python
from time import time

t1 = time()
res = dfSummary(df_wide)
t2 = time()
print(f'{t2 - t1: .2f} sec elapsed')
```

```python
t1 = time()
res = dfSummary(df_wide, num_proc=6) # parallelize into 6 processes
t2 = time()
print(f'{t2 - t1: .2f} sec elapsed')
```
